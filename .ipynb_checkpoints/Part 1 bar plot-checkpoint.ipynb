# Bar plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Number")
plt.title("Pramod")
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Number")
plt.title("Pramod")
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Number")
plt.title("Pramod")
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r",label="Pramod")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Number")
plt.title("Pramod")
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color="r",label="Popularity")
plt.bar(x,z,width=0.4,color="g",label="Popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Number")
plt.title("Pramod")

plt.bar(p,y,width,color="r",label="Popularity")
plt.bar(p1,z,width,color="g",label="Popularity1")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Number")
plt.title("Pramod")

plt.bar(p,y,width,color="r",label="Popularity")
plt.bar(p1,z,width,color="g",label="Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Number")
plt.title("Pramod")

plt.barh(p,y,width,color="r",label="Popularity")
plt.barh(p1,z,width,color="g",label="Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
