# Scatter Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y,color="r")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=["r","y","g","b","r","g","y"]
plt.scatter(x,y,color=color)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=["r","y","g","b","r","g","y"]
size=[10,50,40,20,15,70,90]
plt.scatter(x,y,c=color,s=size)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=["r","y","g","b","r","g","y"]
size=[10,50,40,20,15,70,90]
plt.scatter(x,y,c=color,s=size,alpha=0.4)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=["r","y","g","b","r","g","y"]
size=[100,500,400,200,150,700,900]
plt.scatter(x,y,c=color,s=size,marker="*",edgecolor="g",linewidth=1)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=[10,49,30,29,56,20,30]
size=[600,500,400,800,650,700,900]
plt.scatter(x,y,c=color,s=size,cmap="viridis")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=[10,49,30,29,56,20,30]
size=[600,500,400,800,650,700,900]
plt.scatter(x,y,c=color,s=size,cmap="viridis")
plt.colorbar()
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=[10,49,30,29,56,20,30]
size=[600,500,400,800,650,700,900]
plt.scatter(x,y,c=color,s=size,cmap="viridis")
t=plt.colorbar()
t.set_label("Colorbar",fontsize=20)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
z=[3,2,4,5,1,6,2]
color=[10,49,30,29,56,20,30]
size=[600,500,400,800,650,700,900]
plt.scatter(x,y,c=color,s=size,cmap="viridis")
plt.scatter(x,z,color="r",s=size)
t=plt.colorbar()
t.set_label("Colorbar",fontsize=20)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```
