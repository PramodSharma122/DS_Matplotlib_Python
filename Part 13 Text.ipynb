# Text in matplotlib

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Pramod",fontsize=20)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Pramod",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Pramod",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,5,"java",style="italic",bbox={"facecolor":"red"})

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Pramod",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

plt.text(2,5,"java",style="italic",bbox={"facecolor":"red"})
plt.annotate("python",xy=(3,2),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=100))

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.title("Pramod",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)

# plt.text(2,5,"java",style="italic",bbox={"facecolor":"red"})
# plt.annotate("python",xy=(3,2),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=100))
plt.legend(["up"],loc=9,facecolor="red",edgecolor="yellow",framealpha=0.5,shadow=True)

plt.show()
```
