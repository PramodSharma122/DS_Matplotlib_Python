# Box Plots

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,widths=0.4)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"])
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,patch_artist=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,showmeans=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,whis=3.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,sym="g+")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,boxprops=dict(color="r"))
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b"))
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"),flierprops=dict(markeredgecolor="y"))
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]
y=[x,x1]

plt.boxplot(y,showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]
y=[x,x1]

plt.boxplot(y,labels=["python","c++"] ,showmeans=True,boxprops=dict(color="r"),capprops=dict(color="g"))
plt.show()
```
