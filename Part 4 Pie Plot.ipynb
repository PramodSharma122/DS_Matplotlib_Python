# Pie Plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","python"]

plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","python"]

c=["r","b","g","y"]

plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```
